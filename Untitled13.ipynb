```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# NLTK kütüphanesini indir (Stopwords için)
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

print("1. Veri yükleniyor...")

# GÜNCELLEME BURADA YAPILDI:
# on_bad_lines='skip': Bozuk satırları görmezden gel
# engine='python': Daha esnek okuma motoru kullan
# nrows=5000: Hızlı sonuç görmek için sadece ilk 5000 satırı al
try:
    df = pd.read_csv('train.csv', on_bad_lines='skip', engine='python', nrows=5000)
except TypeError:
    # Eğer pandas sürümü eskiyse 'on_bad_lines' yerine bu çalışır
    df = pd.read_csv('train.csv', error_bad_lines=False, engine='python', nrows=5000)

print(f"Veri başarıyla yüklendi! Boyut: {df.shape}")

# 2. METİN TEMİZLEME (Preprocessing)
def clean_text(text):
    if pd.isna(text): return "" # Boş verileri temizle
    text = str(text).lower() # Küçük harf
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Noktalama işaretlerini sil
    # Stop words (ve, ama, fakat gibi gereksiz kelimeleri) çıkar
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text

print("2. Metinler temizleniyor...")
df['clean_text'] = df['comment_text'].apply(clean_text)

# 3. VEKTÖRLEŞTİRME (TF-IDF)
print("3. Vektörleştirme yapılıyor...")
# max_features=1000: En sık geçen 1000 kelimeyi alıp matris oluşturur
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['clean_text'])

# 4. KÜMELEME (K-Means)
print("4. Kümeleme (K-Means) çalıştırılıyor...")
num_clusters = 3 # Yorumları 3 ana gruba ayırıyoruz
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

df['cluster'] = kmeans.labels_

# 5. SONUÇLARI YAZDIRMA
print("\n--- KÜME ANALİZİ SONUÇLARI ---")
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

for i in range(num_clusters):
    print(f"\n### KÜME {i} ###")
    # Bu kümeyi en iyi anlatan 10 kelime
    top_terms = [terms[ind] for ind in order_centroids[i, :10]]
    print(f"Anahtar Kelimeler: {', '.join(top_terms)}")
    print(f"Yorum Sayısı: {len(df[df['cluster'] == i])}")
    print("Örnek Yorum:", df[df['cluster'] == i]['comment_text'].values[0][:100], "...")

# 6. GÖRSELLEŞTİRME (PCA)
print("\n5. Grafik çiziliyor...")
pca = PCA(n_components=2)
scatter_plot_points = pca.fit_transform(X.toarray())

x_axis = [o[0] for o in scatter_plot_points]
y_axis = [o[1] for o in scatter_plot_points]

plt.figure(figsize=(12, 8))
sns.scatterplot(x=x_axis, y=y_axis, hue=df['cluster'], palette='viridis', s=50)
plt.title('Clustering of Toxic Comments using K-Means')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster No')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

print("En ideal grup sayısı hesaplanıyor (Bu işlem 10 kez çalışacak)...")

wcss = [] # Within-Cluster Sum of Square (Hata kareler toplamı)
cluster_range = range(1, 11) # 1'den 10'a kadar küme sayılarını dene

for i in cluster_range:
    # Her döngüde farklı küme sayısıyla modeli kur
    kmeans_test = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans_test.fit(X)
    # Inertia değerini (hatayı) listeye ekle
    wcss.append(kmeans_test.inertia_)
    print(f"Küme Sayısı {i} için hesaplama tamamlandı.")

# Grafiği Çiz
plt.figure(figsize=(10, 5))
plt.plot(cluster_range, wcss, marker='o', linestyle='--', color='b')
plt.title('Elbow Method')
plt.xlabel('Cluster Count (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.cluster import KMeans

# 1. RETRAINING THE MODEL WITH THE IDEAL NUMBER (K=6)
print("Retraining model with K=6...")
ideal_k = 6
kmeans_final = KMeans(n_clusters=ideal_k, random_state=42)
kmeans_final.fit(X)
df['cluster'] = kmeans_final.labels_

# 2. GENERATING WORD CLOUDS
print("Preparing word clouds...")

plt.figure(figsize=(20, 10))

for i in range(ideal_k):
    # Combining all comments belonging to that cluster into a single long text
    cluster_text = " ".join(df[df['cluster'] == i]['clean_text'])

    # Generate the word cloud
    wc = WordCloud(width=800, height=400, background_color='white', max_words=50).generate(cluster_text)

    # Place into the subplot
    plt.subplot(2, 3, i+1) # 2 rows, 3 columns grid
    plt.imshow(wc, interpolation='bilinear')
    plt.title(f"Cluster {i} (Size: {len(df[df['cluster'] == i])})", fontsize=16)
    plt.axis('off')

plt.tight_layout()
plt.show()

# Statistical Summary
print("\n--- INSIGHTS ON GROUP CONTENTS ---")
order_centroids = kmeans_final.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

for i in range(ideal_k):
    top_terms = [terms[ind] for ind in order_centroids[i, :5]]
    print(f"Cluster {i} Dominant Words: {', '.join(top_terms)}")
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns

# Download VADER lexicon
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

print("Performing sentiment analysis (Calculating average score for each cluster)...")

# Calculate 'compound' sentiment score for each comment
# Returns a value between -1 (Very Negative) and +1 (Very Positive)
df['sentiment_score'] = df['clean_text'].apply(lambda x: sia.polarity_scores(str(x))['compound'])

# Group by cluster and calculate the mean
cluster_sentiment = df.groupby('cluster')['sentiment_score'].mean().sort_values()

print("\n--- CLUSTER SENTIMENT SCORES (From Most Negative to Positive) ---")
print(cluster_sentiment)

# Visualization
plt.figure(figsize=(10, 5))
sns.barplot(x=cluster_sentiment.index, y=cluster_sentiment.values, palette='coolwarm')
plt.title('Average Sentiment Score of Clusters (-1: Negative, +1: Positive)')
plt.ylabel('Average Sentiment Score')
plt.xlabel('Cluster Number')
plt.axhline(0, color='black', linestyle='--') # Draw a reference line at 0
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

print("Running Anomaly Detection (Isolation Forest)...")

# Isolation Forest identifies points that are far from dense regions of data.
# contamination=0.01 -> We are asking to find the weirdest 1% of the data.
iso_forest = IsolationForest(contamination=0.01, random_state=42)
anomalies = iso_forest.fit_predict(X.toarray())

# -1: Anomaly, 1: Normal
df['anomaly'] = anomalies

# Filter only the anomalies (-1)
outliers = df[df['anomaly'] == -1]

print(f"\nTotal {len(outliers)} anomalies (unusual comments) detected.")
print("-" * 50)

# Let's print the top 3 weirdest comments
for i in range(3):
    try:
        row = outliers.iloc[i]
        print(f"\n--- ANOMALY #{i+1} (Cluster: {row['cluster']}) ---")
        # Truncate if the comment is too long
        text_preview = row['comment_text'][:300] + "..." if len(str(row['comment_text'])) > 300 else row['comment_text']
        print(f"Content: {text_preview}")
    except:
        pass

print("-" * 50)
print("These comments are 'outlier' data that do not fit standard language patterns.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import IsolationForest

# STEP 1: Cleaning Bot/System Messages
# Common words in Wikipedia templates
bot_keywords = [
    "speedily deleted", "tag has been placed", "copyright violation",
    "blocked", "IP address", "unsigned", "automatically", "bot",
    "jpg", "png", "image:", "file:", "utc"
]

# Creating regex pattern
pattern = '|'.join(bot_keywords)

# NOTE: We are using the 'comment_text' column here
# Selecting rows where these words do NOT appear (~)
df_clean = df[~df['comment_text'].str.contains(pattern, case=False, na=False)].copy()

print(f"Original Data Count: {len(df)}")
print(f"Cleaned Data Count: {len(df_clean)}")
print(f"Removed Bot/System Messages Count: {len(df) - len(df_clean)}")

# STEP 2: Vectorization (With cleaned data)
# Using TF-IDF to convert words into numbers for the model
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X_clean = vectorizer.fit_transform(df_clean['comment_text'])

# STEP 3: Re-running Isolation Forest
# contamination=0.01 -> Treat the weirdest 1% of data as anomalies
iso_forest = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
preds = iso_forest.fit_predict(X_clean)

# Save results (-1: Anomaly, 1: Normal)
df_clean['new_anomaly'] = preds

# STEP 4: Inspecting New Anomalies
new_anomalies = df_clean[df_clean['new_anomaly'] == -1]

print("-" * 30)
print(f"Number of New Anomalies: {len(new_anomalies)}")
print("-" * 30)

# Print the first 3 new anomalies
for index, row in new_anomalies.head(3).iterrows():
    print(f"--- NEW ANOMALY (ID: {row['id']}) ---")
    # Show only the first 300 characters if too long
    print(row['comment_text'][:300] + "...")
    print("\n")
```

```python
# How many of the anomalies are actually labeled as 'Toxic'?
anomaly_subset = df_clean[df_clean['new_anomaly'] == -1]

# Getting the count of rows where the 'toxic' column is 1
toxic_count = anomaly_subset['toxic'].sum()
total_anomalies = len(anomaly_subset)

print(f"Total Number of Anomalies: {total_anomalies}")
print(f"Count of 'Toxic' Labeled Among These: {toxic_count}")
print(f"Ratio: {(toxic_count / total_anomalies) * 100:.2f}%")

# What type of toxicity was caught? (Let's look at the means)
print("\n--- Characteristics of Detected Anomalies ---")
print(anomaly_subset[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].mean())
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.ensemble import IsolationForest
import numpy as np
import pandas as pd

# STEP 1: Load Deep Learning Model
# 'all-MiniLM-L6-v2' is both fast and very effective.
print("Loading model...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# STEP 2: Convert Texts to "Semantic Vectors" (Encoding)
# This process may take 1-2 minutes depending on data size.
# df_clean is your dataset cleaned of bots.
print("Converting comments to vectors...")
embeddings = model.encode(df_clean['comment_text'].tolist(), show_progress_bar=True)

# STEP 3: Run Isolation Forest on Vectors
# contamination=0.01 -> Find the weirdest 1%
iso_forest_bert = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
bert_preds = iso_forest_bert.fit_predict(embeddings)

# Save results
df_clean['bert_anomaly'] = bert_preds

# STEP 4: Inspect Results
bert_anomalies = df_clean[df_clean['bert_anomaly'] == -1]

print("\n" + "="*40)
print(f"Number of Anomalies Found by BERT: {len(bert_anomalies)}")
print("="*40)

# Check Toxicity Ratio
toxic_count = bert_anomalies['toxic'].sum()
print(f"Count of 'Toxic' Labeled Among These: {toxic_count}")
print(f"BERT Success Rate (Toxic Capture): {(toxic_count / len(bert_anomalies)) * 100:.2f}%")
print("-" * 30)

# Print Examples
print("\n--- Top 3 Weirdest Comments Found by BERT ---")
for index, row in bert_anomalies.head(3).iterrows():
    print(f"\n[ID: {row['id']}] (Toxic: {row['toxic']})")
    print(row['comment_text'][:300] + "...")
```

```python
from sklearn.cluster import KMeans

# STEP 1: K-Means Model (Let's split data into 5 main groups)
# Why 5? (1. Normal Chat, 2. Politics/Religion, 3. Tech/Wikipedia, 4. Conflict/Toxic, 5. Others)
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)

# Train model on vectors
df_clean['cluster'] = kmeans.fit_predict(embeddings)

# STEP 2: Which Cluster is "Toxic"?
# Let's look at the average 'toxic' rate for each cluster
print("-" * 40)
print("CLUSTER ANALYSIS (Which group is more dangerous?)")
print("-" * 40)

# Calculate the mean of 'toxic' column by groups
cluster_stats = df_clean.groupby('cluster')['toxic'].mean().sort_values(ascending=False)
print(cluster_stats)

# STEP 3: Examples from the Most "Toxic" Cluster
most_toxic_cluster_id = cluster_stats.index[0] # Cluster ID with the highest rate
toxic_cluster_data = df_clean[df_clean['cluster'] == most_toxic_cluster_id]

print("\n" + "="*40)
print(f"MOST DANGEROUS GROUP: Cluster {most_toxic_cluster_id}")
print(f"Number of comments in this group: {len(toxic_cluster_data)}")
print("="*40)

print("3 random examples from this cluster:")
for idx, row in toxic_cluster_data.sample(3).iterrows():
    print(f"\n[Toxic: {row['toxic']}]")
    print(row['comment_text'][:300] + "...")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# 1. Dimensionality Reduction (from 384 dimensions -> 2 dimensions)
# This allows us to plot on X and Y axes.
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)

# 2. Add data to DataFrame
df_clean['x'] = reduced_embeddings[:, 0]
df_clean['y'] = reduced_embeddings[:, 1]

# 3. Visualization
plt.figure(figsize=(12, 8))

# Color palette: Color by clusters
# hue='cluster' -> Separate colors by cluster
# style='toxic' -> Separate shapes (Circle/Cross) by toxicity
sns.scatterplot(
    data=df_clean,
    x='x',
    y='y',
    hue='cluster',
    style='toxic',
    palette='tab10',
    alpha=0.7,
    s=60 # Dot size
)

plt.title('Semantic Map of Comments (K-Means + BERT)', fontsize=15)
plt.xlabel('PCA Component 1 (Semantic Axis 1)')
plt.ylabel('PCA Component 2 (Semantic Axis 2)')
plt.legend(title='Cluster / Toxic', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Helper Function: Find the most frequent words
def get_top_n_words(corpus, n=15):
    vec = CountVectorizer(stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

# 2. Separate Clusters (Green vs Orange)
# In the plot, Cluster 2 (Green) appeared toxic, Cluster 1 (Orange) appeared clean.
# If the colors are different in your plot, you can change the numbers here accordingly.
green_cluster_text = df_clean[df_clean['cluster'] == 2]['comment_text']
orange_cluster_text = df_clean[df_clean['cluster'] == 1]['comment_text']

# 3. Count Words
top_words_green = get_top_n_words(green_cluster_text, 20)
top_words_orange = get_top_n_words(orange_cluster_text, 20)

# Convert data to DataFrame for plotting
df_green = pd.DataFrame(top_words_green, columns=['Word', 'Freq'])
df_green['Cluster'] = 'Green Region (Toxic/Conflict)'

df_orange = pd.DataFrame(top_words_orange, columns=['Word', 'Freq'])
df_orange['Cluster'] = 'Orange Region (Info/Clean)'

# 4. Visualization (Side-by-Side Bar Chart)
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

sns.barplot(data=df_green, x='Freq', y='Word', ax=axes[0], palette='Reds_r')
axes[0].set_title('Green Region (Cluster 2) Words', fontsize=15)

sns.barplot(data=df_orange, x='Freq', y='Word', ax=axes[1], palette='Oranges_r')
axes[1].set_title('Orange Region (Cluster 1) Words', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import joblib
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
import plotly.express as px

# --- AYARLAR ---
st.set_page_config(page_title="Toksik Yorum Dedektörü", layout="wide")

# --- BAŞLIK VE AÇIKLAMA ---
st.title("🕵️‍♂️ Yapay Zeka Tabanlı Troll Dedektörü")
st.markdown("""
Bu sistem, yorumlardaki kelimelere değil, **cümlenin anlamına** odaklanır.
Arka planda **BERT** ve **K-Means Clustering** çalışmaktadır.
""")

# --- MODELLERİ YÜKLEME ---
@st.cache_resource # Modelleri önbelleğe al ki her seferinde tekrar yüklemesin
def load_models():
    kmeans = joblib.load('kmeans_model.pkl')
    pca = joblib.load('pca_model.pkl')
    toxic_id = joblib.load('toxic_cluster_id.pkl')
    bert_model = SentenceTransformer('all-MiniLM-L6-v2')
    return kmeans, pca, toxic_id, bert_model

kmeans, pca, toxic_id, model = load_models()

# --- YAN MENÜ (SIDEBAR) ---
st.sidebar.header("Nasıl Çalışır?")
st.sidebar.info("""
1. Yorumunuz BERT ile vektöre çevrilir.
2. Vektör uzayındaki konumu bulunur.
3. Eğer 'Kavga Bölgesi'ne (Cluster 2) düşerse uyarı verilir.
""")

# --- ANA EKRAN ---
user_input = st.text_area("Test etmek istediğiniz İngilizce yorumu girin:", height=100, placeholder="Örnek: You are so stupid and I hate this article!")

if st.button("Analiz Et"):
    if user_input:
        with st.spinner('Yapay Zeka düşünüyor...'):
            # 1. Vektöre Çevir
            embedding = model.encode([user_input])

            # 2. Hangi Kümeye Ait?
            cluster_pred = kmeans.predict(embedding)[0]

            # 3. Koordinatları Bul (Görselleştirme için)
            coords = pca.transform(embedding)
            x_pos, y_pos = coords[0][0], coords[0][1]

            # --- SONUÇ EKRANI ---
            col1, col2 = st.columns([1, 2])

            with col1:
                st.subheader("Sonuç:")
                if cluster_pred == toxic_id:
                    st.error(f"🚨 TOKSİK TESPİT EDİLDİ!\n(Küme: {cluster_pred})")
                    st.write("Bu yorum, veri setindeki 'kavga ve hakaret' bölgesine düştü.")
                else:
                    st.success(f"✅ GÜVENLİ YORUM\n(Küme: {cluster_pred})")
                    st.write("Bu yorum temiz veya teknik bilgi içeriyor.")

            with col2:
                # Basit bir radar/konum grafiği (Plotly ile)
                # Not: Gerçek veriyi yüklemek uzun sürer, temsili bir görsel yapıyoruz
                df_visual = pd.DataFrame({
                    'x': [x_pos],
                    'y': [y_pos],
                    'label': ['Sizin Yorumunuz'],
                    'color': ['Toksik' if cluster_pred == toxic_id else 'Güvenli']
                })

                fig = px.scatter(df_visual, x='x', y='y', color='color',
                                color_discrete_map={'Toksik':'red', 'Güvenli':'green'},
                                title="Yorumunuzun Vektör Uzayındaki Konumu (Temsili)",
                                size=[20], range_x=[-0.5, 0.5], range_y=[-0.5, 0.5])

                # Merkezi (0,0) göster
                fig.add_shape(type="circle", xref="x", yref="y", x0=-0.1, y0=-0.1, x1=0.1, y1=0.1, line_color="LightSkyBlue")

                st.plotly_chart(fig, use_container_width=True)

    else:
        st.warning("Lütfen bir metin girin.")
```
